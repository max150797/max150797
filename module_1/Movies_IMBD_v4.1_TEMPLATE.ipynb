# Предобработка

```python
answers = {} # создадим словарь для ответов
```

```python
def func_split(string):         # функция для преобразования строки с разделителем "|" в список
    return string.split('|')


def func_date(string):         # функция для преобразования строки с разделителем "\" в список
    return string.split('/')


data['profit'] = data['revenue'] - data['budget'] # добавим столбец прибыли

# Ниже добавим колонки, которые будут содержать в себе преобразованные в список данные
data['list_of_cast'] = data.cast.apply(func_split) 

data['list_of_genres'] = data.genres.apply(func_split) 

data['list_of_director'] = data.director.apply(func_split)

data['list_of_production_companies'] = data.production_companies.apply(func_split)

data['release_month'] = data.release_date.apply(func_date) #колонка месяца выхода фильма
data['release_month'] = data.release_month\
.apply(lambda x: int(x[0])) # преобразуем формат str в int

# добавим столбец с длиной названий фильмов
data['title_length'] = data.original_title.apply(lambda title: len(title))
# добавим столбец с длиной описания к фильму
data['overview_length'] = data.overview.apply(lambda overview: len(overview))
```
