##### На гистограмме видно, что присутствует выброс, поскольку данный показатель изменяется от 1 до 5. Можно предположить, что, как и в предыдущий раз была допущена ошибка при переносе данных. Предлагается исправить -1 на 1, т.к. даже если у школьника очень плохие отношения с родителями и он решил указать (-1), то в любом случае это относится к категории 1-очень плохо.

```python
# Заменим единственный выброс -1 на 1
students.famrel = students.famrel.replace(-1, 1)
```

```python
# Проверим
students.famrel.hist()
display(students.famrel.describe())
display(students.famrel.value_counts(dropna = False).reset_index())
```

##### В данном случае гистограмма является корректной. Можно сказать, что у большинства школьников отношения с родителями хорошие и очень хорошие (25-ый квартиль 4.0, 75-ый квартиль 5.0)

```python
students.freetime.hist()
display(students.freetime.describe())
display(students.freetime.value_counts(dropna = False).reset_index())
```

##### Данная гистограмма показывает, что показатель свободного времени у  школьников после занятий оценивается как средний.

```python
students.goout.hist()
display(students.goout.describe())
display(students.goout.value_counts(dropna = False).reset_index())
```

###### Поскольку на данной гистограмме медианное значение почти совпадает со средним, то можно сказать, что распределение является нормальным с медианной равной 3 (большинство школьников проводят с друзьями среднее количество времени)

```python
students.health.hist()
display(students.health.describe())
display(students.health.value_counts(dropna = False).reset_index())
```
